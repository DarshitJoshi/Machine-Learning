## Support Vector Classifier

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
##Lets create synthetic datapoints
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=1,n_redundant=0)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)[0]
```

```python
df = pd.DataFrame(X, columns=['Feature1', 'Feature2'])
df['Label'] = y
```

```python
# Create scatter plot
sns.scatterplot(data=df, x='Feature1', y='Feature2', hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

```python
## Prediction
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
rbf = SVC(kernel='rbf')
```

```python
rbf.fit(X_train,y_train)
```

```python
## Prediction
y_pred1 = rbf.predict(X_test)
```

```python
print(classification_report(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
```

```python
polynomial = SVC(kernel='poly')
polynomial.fit(X_train,y_train)
y_pred2 = polynomial.predict(X_test)
print(classification_report(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
```

Polynomial transformation will not work because it is giving less accuracy

```python
sigmoid = SVC(kernel='sigmoid')
sigmoid.fit(X_train,y_train)
y_pred3 = sigmoid.predict(X_test)
print(classification_report(y_test,y_pred3))
print(confusion_matrix(y_test,y_pred3))
```

 rbf has given best accuracy so i will use it

```python
## HyperParameter tuning with SVC
```

```python
from sklearn.model_selection import GridSearchCV

# define parameter range
param_grid = {
    'C':[0.1,1,10,100,1000],
    'gamma':[1,0.1,0.01,0.001,0.001],
    'kernel':['rbf']
}
```

```python
grid=GridSearchCV(SVC(),param_grid=param_grid,refit=True,cv = 5,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
#prediction
y_pred4 = grid.predict(X_test)
print(classification_report(y_test,y_pred4))
print(confusion_matrix(y_test,y_pred4))
```

Accuracy improved after selecting right C,gamma,kernel values
